## 3D Scatter Plots

Most plots use only **two dimensions** (x and y).

However, sometimes we want to visualize **three variables at once**. A 3D plot allows us to represent data using:

- X axis
- Y axis
- Z axis

### Example

Suppose we measure:

- Study hours
- Sleep hours
- Exam scores

```python
study = [2,3,4,5,6,7,8]
sleep = [7,6,6,5,7,6,5]
score = [60,65,70,72,80,85,90]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(study, sleep, score)

ax.set_xlabel("Study Hours")
ax.set_ylabel("Sleep Hours")
ax.set_zlabel("Exam Score")

plt.title("3D Scatter Plot Example")

plt.show()
```

#### This is a bad plot as generally you would want to view any scatter plot to understand the relationship in 2 dimensions i.e., x & y. You represent the 3rd relationship by giving a color scheme to the x,y plotted points.

3D plots help visualize relationships between **three variables at the same time**.

However, they should be used carefully because they can sometimes be harder to interpret than 2D plots.

## Bubble Charts

A **bubble chart** is a variation of a scatter plot.

In addition to:

- X position
- Y position

it also uses **bubble size** to represent a third variable.

### Example

Suppose we want to compare countries using:

- GDP
- Life expectancy
- Population size

```python
gdp = [1000,2000,3000,4000,5000]
life = [60,65,70,75,80]
population = [10,20,30,40,50]

plt.scatter(gdp, life, s=np.array(population)*20, alpha=0.5)

plt.xlabel("GDP per Capita")
plt.ylabel("Life Expectancy")
plt.title("Bubble Chart Example")

plt.show()
```

In this chart:

- **X axis** shows GDP
- **Y axis** shows life expectancy
- **Bubble size** represents population

## Animated Charts

Animations allow us to show **how data changes over time**.

This is useful for:

- economic data
- population growth
- climate data
- scientific simulations

```python
data = px.data.gapminder()

fig = px.scatter(
    data,
    x="gdpPercap",
    y="lifeExp",
    animation_frame="year",
    animation_group="country",
    size="pop",
    color="continent",
    hover_name="country",
    log_x=True
)

fig.show()
```

This animation shows how:

- GDP
- life expectancy
- population

change across countries over time.

## Network Graphs

Network graphs visualize **relationships between objects**.

They are commonly used in:

- social networks
- transportation systems
- communication networks
- biology

Network graphs contain:

**Nodes**
- represent entities

**Edges**
- represent connections

```python
G = nx.Graph()

G.add_edges_from([
    ("Alice","Bob"),
    ("Alice","Charlie"),
    ("Bob","David"),
    ("Charlie","David"),
    ("David","Eva")
])

nx.draw(G, with_labels=True)

plt.title("Simple Network Graph")

plt.show()
```

Network graphs help us understand **how different entities are connected**.

## Pair Plots

Pair plots allow us to visualize relationships between **multiple variables at once**.

They create scatter plots for every pair of variables in a dataset.

This is useful during **exploratory data analysis**.

```python
df = sns.load_dataset("iris")

sns.pairplot(df, hue="species")

plt.show()
```

Pair plots help reveal:

- correlations
- clusters
- group differences

## Multi-Variable Time Series

Sometimes we want to compare **multiple variables over time**.

For example:

- temperature
- rainfall
- river discharge

```python
days = np.arange(1,8)

temp = [20,21,19,22,23,24,25]
rain = [5,3,6,2,0,1,0]

plt.plot(days, temp, label="Temperature")
plt.plot(days, rain, label="Rainfall")

plt.xlabel("Day")
plt.ylabel("Value")

plt.title("Weather Variables Over Time")

plt.legend()

plt.show()
```

## Interactive Visualization

Interactive plots allow users to:

- zoom
- hover over points
- filter data

These features help users explore complex datasets.

```python
df = px.data.iris()

fig = px.scatter(
    df,
    x="sepal_width",
    y="sepal_length",
    color="species"
)

fig.show()
```

## Choosing the Right Advanced Plot

| Plot Type | Best Use |
|-----------|-----------|
| 3D Plot | Visualizing three variables |
| Bubble Chart | Showing magnitude with point size |
| Animated Chart | Showing change over time |
| Network Graph | Relationships and connections |
| Pair Plot | Exploring many variables |
| Interactive Plot | Data exploration |

Always remember:

A visualization should **help people understand the data**, not make it more confusing.

## What is a Variable?

A **variable** is like a labeled box that holds data.  
You give the box a name and put something inside it.

Python is very flexible — you don’t need to tell it what kind of data you’re storing.

```python
# Let's create our first variables
name = "Alex"          # text (called a "string")
age = 23               # whole number (called an "integer")
height_cm = 175.5      # number with decimal (called a "float")
is_student = True      # True or False (called a "boolean")

# Print them out so we can see what's inside the boxes
print("Name:", name)
print("Age:", age)
print("Height:", height_cm, "cm")
print("Is a student?", is_student)
```

## How Jupyter Notebooks Store Variables

Here’s the most important thing to understand:

> **Variables live in the Kernel** (a background Python process).

When you run a cell, the code is sent to the kernel and the variables are stored **in memory**.  
They stay there even after you scroll away.

This means:
- You can use a variable in any cell below (or even above!)
- If you change a variable in one cell, **every other cell that uses it will see the new value** — but only after you re-run that cell.

## Let's see this in action

Run the cells below **in order** first.  
Then we’ll try running them out of order to see what happens.

```python
# Cell A – Create the variable
score = 85
print("Initial score:", score)
```

```python
# Cell B – Update the variable
score = score + 10
print("Updated score:", score)
```

```python
# Cell C – Use the variable again
print("Current score is still:", score)
```

## Try this now:

1. Run Cell A → Cell B → Cell C (you should see 85 → 95)
2. Go back to **Cell A** and change `score = 85` to `score = 200`
3. Run **only Cell A** again
4. Then run **Cell C** (without running Cell B)

What do you see?

→ The score is now 200!  
Cell B was never run again, but Cell C still shows the new value because the kernel remembers the latest version of `score`.

This is why the **order you run cells** matters more than the order they appear in the notebook.

## Updating Variables (very common in data science)

You will do this a lot when cleaning data or building models.

```python
# Start with a list of numbers (we'll use this a lot later)
data_points = [10, 20, 30, 40]

# Update the variable
data_points = data_points + [50]   # add a new number

print("Updated list:", data_points)
print("Number of points:", len(data_points))
```

## What if you want to "delete" a variable?

Use the `del` command.

```python
# Delete a variable
del score

# Try to use it now
try:
    print(score)
except NameError:
    print("Error: 'score' no longer exists!")
```

## Best Practices for Workshop & Real Projects

1. **Always run from the top**  
   → Kernel → Restart & Run All (this clears everything and runs every cell in order)

2. If something feels “weird”, restart the kernel:  
   → Kernel → Restart (or Restart & Clear Outputs)

3. Name your variables clearly (e.g., `customer_age`, `sales_data`)

4. Add comments in your code — future you will thank you!


## 1. Scatter Plots

A **scatter plot** shows the relationship between two numerical variables.

Each point on the chart represents one observation.

Scatter plots are useful for identifying:

- Relationships between variables
- Trends
- Clusters
- Outliers

### Example

Suppose we want to examine the relationship between:

- Hours studied
- Exam score

```python
hours = [1,2,3,4,5,6,7,8]
scores = [50,55,60,65,70,78,85,90]

plt.scatter(hours, scores)

plt.title("Hours Studied vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")

plt.show()
```

### Interpretation

From this plot we can see that:

As study hours increase, exam scores tend to increase.

This suggests a **positive relationship** between study time and exam performance.

## 2. Heatmaps

A **heatmap** shows data values using colors.

Instead of numbers, colors represent the intensity or magnitude of values.

Heatmaps are commonly used for:

- Correlation matrices
- Density patterns
- Geographic data
- Activity over time

### Example

We will create a **correlation heatmap** using a small dataset.

```python
data = {
    "Math":[85,78,90,88,76],
    "Science":[92,81,95,89,80],
    "English":[75,72,78,70,68],
    "History":[80,77,85,82,74]
}

df = pd.DataFrame(data)

corr = df.corr() # pre-defined function of python, which measures the relation between 2 variables.

sns.heatmap(corr, annot=True)

plt.title("Subject Score Correlations")

plt.show()
```

### Interpretation

Heatmaps allow us to quickly see relationships.

- Darker colors usually represent **stronger relationships**
- Lighter colors represent **weaker relationships**

This makes it easier to interpret complex datasets.

## 3. Pie Charts

A **pie chart** shows how a whole is divided into parts.

Each slice represents a proportion of the total.

Pie charts are best used when:

- The number of categories is small
- You want to show proportions or percentages

### Example

Suppose we want to visualize how students spend their study time.

```python
labels = ["Reading", "Practice", "Videos", "Group Study"]
sizes = [35, 40, 15, 10]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")

plt.title("Student Study Time Distribution")

plt.show()
```

### Important Note

Pie charts should be used carefully.

If there are too many categories, the chart becomes difficult to read. In many cases, a **bar chart is a better alternative**.

## 4. Treemaps

A **treemap** displays hierarchical data using nested rectangles.

The size of each rectangle represents a value.

Treemaps are useful for visualizing:

- Budget allocations
- File system storage
- Company market share
- Population distributions

### Example

Let us visualize how a company distributes its budget across departments.

```python
departments = ["Marketing", "Research", "Operations", "HR", "IT"]
budget = [40, 25, 20, 10, 5]

squarify.plot(sizes=budget, label=departments, alpha=0.8)

plt.title("Company Budget Allocation")

plt.axis("off")

plt.show()
```

### Interpretation

In a treemap:

- Larger rectangles represent **larger values**
- Smaller rectangles represent **smaller values**

This allows us to quickly compare proportions across categories.

## 5. Choropleth Maps

A **choropleth map** displays geographic data using colors.

Each region on the map is shaded based on a data value.

Choropleth maps are widely used for:

- Population density
- Election results
- Economic indicators
- Public health statistics

### Example

We will create a simple choropleth map showing population values by country.

More about gapminder function can be found here. Pre-loaded population data is being loaded below. https://www.kaggle.com/code/jhossain/explore-the-gapminder-dataset-with-plotly-express

```python
data = px.data.gapminder().query("year == 2007")

fig = px.choropleth(
    data,
    locations="iso_alpha",
    color="pop",
    hover_name="country",
    title="World Population by Country (2007)"
)

fig.show()
```

### Why Choropleths Are Useful

Choropleth maps allow us to visualize **geographic patterns** quickly.

For example, we might discover:

- Regions with higher population
- Areas with higher disease rates
- Countries with higher GDP

## Choosing the Right Visualization

Different visualizations answer different questions.

| Visualization | Best Used For |
|---------------|---------------|
| Scatter Plot | Relationship between two numeric variables |
| Heatmap | Patterns and correlations |
| Pie Chart | Showing proportions |
| Treemap | Hierarchical or proportional data |
| Choropleth | Geographic patterns |

Choosing the correct visualization is an important skill in data science.

1. Understand patterns and trends

2. Detect outliers and anomalies

3. Communicate insights clearly

4. Explore data before modeling

**In this tutorial, we will learn how to create simple plots in Python using Matplotlib inside a Jupyter Notebook.**

# Step 1: Importing Required Libraries

### Importing Libraries

Python uses **libraries** to perform specialized tasks.

For plotting, we primarily use:
- **matplotlib** → creating plots and charts
- **numpy** → working with numbers and arrays

```python
import matplotlib.pyplot as plt
import numpy as np
```

* `plt` is a common shorthand for matplotlib.pyplot

* `np` is a common shorthand for numpy

# Step 2: Your First Line Plot

## Creating Our First Plot

A **line plot** shows how a value changes over time or across observations.

```python
# Create some data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot
plt.plot(x, y)

# Display the plot
plt.show()
```

## Step 3: Adding Labels and a Title

### Adding Labels and a Title

Good plots always have:
- A title
- Labeled axes

```python
plt.plot(x, y)

plt.title("Simple Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.show()
```

# Step 4: Customizing Plot Appearance

## Customizing Plots

We can change colors, line styles, and markers to improve readability.

```python
plt.plot(x, y, color="blue", linestyle="--", marker="o")

plt.title("Customized Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.show()
```

# Step 5: Plotting Multiple Lines
## Plotting Multiple Lines

We can display more than one dataset on the same plot.

```python
print(x)
print(y)
```

```python
y2 = [1, 2, 3, 4, 5] # we need another function

plt.plot(x, y, label="y = 2x")
plt.plot(x, y2, label="y = x")

plt.title("Multiple Lines on One Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()

plt.show()
```

# Step 6: Using NumPy for Data Generation
## Using NumPy to Generate Data

NumPy helps us create numerical data easily.

```python
x = np.linspace(0, 10, 50)
y = np.sin(x) # sin function of randomly generated x values.

plt.plot(x, y)

plt.title("Sine Wave")
plt.xlabel("X")
plt.ylabel("sin(X)")

plt.show()
```

## The standard usage of numpy.linspace(start, stop, num):
`start (0)`: The starting value of the sequence.

`stop (10)`: The end value of the sequence (included by default).

`num (50)`: The number of samples to generate.

# Step 7: Creating a Scatter Plot
## Scatter Plots

Scatter plots show individual data points and are useful for exploring relationships.

```python
x = np.random.rand(50) # random 50 points
y = np.random.rand(50)

plt.scatter(x, y)

plt.title("Scatter Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.show()
```

# Step 8: Plotting Data from a Table (Pandas)
## Plotting Data Using Pandas

In real data science workflows, data often comes in tables.

```python
import pandas as pd # used to handle tables, we can read files such as .csv/excel files as well with this package.

# Create a simple DataFrame (dictionary)
# imagine year, sales as the columns
data = {
    "Year": [2018, 2019, 2020, 2021, 2022],
    "Sales": [200, 250, 300, 280, 350]
}

df = pd.DataFrame(data)

# Plot using pandas
df.plot(x="Year", y="Sales", kind="line")

plt.title("Sales Over Time")
plt.ylabel("Sales")
plt.show()
```

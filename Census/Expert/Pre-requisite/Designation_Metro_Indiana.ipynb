## Read the Shp File that has the Metro Info

```python
gdf1 =  gpd.read_file("Metro_Indiana/Indiana_Tract_Metro.shp")
```

```python
gdf1
```

```python
gdf1.columns.tolist()
```

```python
# Create a dictionary from gdf1 for faster lookup
metro_dict = dict(zip(gdf1['GEOID'], gdf1['NAME_1']))
```

```python
# Add the new columns to gdf
gdf['Metro'] = gdf['GEOID'].apply(lambda x: 'Yes' if x in metro_dict else 'No')
gdf['NAME_1'] = gdf['GEOID'].apply(lambda x: metro_dict.get(x, 'Not Applicable'))
```

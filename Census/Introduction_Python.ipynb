## 8. Lists (Ordered Collections)

```python
numbers = [10, 20, 30, 40]
names = ["Alice", "Bob", "Charlie"]
```

#### Accessing Elements

```python
numbers[0]      # First element
```

##### Numbering in python starts from 0. As in 0,1,2,3,4, .... . -1 indicates the last element in a list.  

```python
numbers[-1]     # Last element
```

```python
numbers[-2]
```

#### Modifying Lists

```python
numbers.append(50) # adding 50 to list
numbers.remove(20) # removing 20 from list
numbers # print new list
```

## 10. Conditional Statements (if / else)

```python
score = 85 # for example
```

```python
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
else:
    print("Grade: C or lower")
```

```python
## try it yourself! update score value and see the change.
```

## 11. Loops

```python
for num in [1, 2, 3, 4, 5]:
    print(num)
```

#### Printing the range. This prints the index value.

```python
for i in range(5):
    print(i)
```

## 12. Functions - Great for reusing/streamlining the code.

```python
def greet(name):
    return f"Hello, {name}!"
```

```python
greet("Data Scientist")
```

### greet is the function here. Here `name` is the parameter that we update calling the function `greet`. Name gets updated and returns the updated print statement.

# Good vs Bad Data Visualization in Python

## Introduction

Data visualization is one of the most important skills in data science. A good plot can make complex information easy to understand, while a poorly designed plot can confuse people or even lead to incorrect conclusions.

In this tutorial we will learn how to:

- Recognize **bad charts**
- Improve **titles and labels**
- Fix **axis scales**
- Choose **better chart types**
- Make plots **clear and readable**

We will use the Python library **Matplotlib**, which is one of the most widely used plotting libraries.

This tutorial is designed for **beginners who are new to Python and data visualization**.

---

## What Makes a Plot Bad?

Some common problems with charts include:

- Missing titles
- Missing axis labels
- Misleading axis scales
- Too many colors
- Using the wrong chart type
- Hard-to-read formatting

Throughout this notebook we will look at **examples of bad plots and improve them step by step.**

---

## Import Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Example Dataset

First we will create a simple dataset. Imagine we recorded the number of visitors to a website during one week.

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
visitors = [120, 135, 150, 160, 200, 300, 280]
```

## Example 1: A Bad Plot

The following plot is not very helpful.

Problems with this plot:

- No title
- No axis labels
- Readers do not know what the data represents

Let's look at the bad version first.

```python
plt.plot(days, visitors)
plt.show()
```

### Why is this plot bad?

If someone looks at this chart, they might ask:

- What is being measured?
- What do the numbers represent?
- What does the x-axis represent?

Without labels and titles, the plot lacks context.

## Fixing the Plot

We can improve the plot by adding:

- A clear title
- Axis labels
- Better formatting

```python
plt.plot(days, visitors)

plt.title("Website Visitors Over One Week")
plt.xlabel("Day of the Week")
plt.ylabel("Number of Visitors")

plt.show()
```

Now the chart is much easier to understand.

A good title answers the question:

**"What story is this chart telling?"**

## Example 2: Misleading Axis Scale

Sometimes charts are misleading because of how the axis scale is set.

In the example below, the y-axis is truncated. This exaggerates the differences between values.

```python
sales = [500, 520, 530, 540, 560]
months = ["Jan", "Feb", "Mar", "Apr", "May"]

plt.bar(months, sales)
plt.ylim(490, 570)

plt.title("Company Sales")
plt.show()
```

### Why is this misleading?

The differences between months appear very large because the y-axis starts at **490 instead of 0**.

This exaggerates the visual differences.

Good visualization practice usually starts bar charts at **zero**.

## Fixing the Axis Scale

```python
plt.bar(months, sales)

plt.title("Monthly Company Sales")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.ylim(0, 600)

plt.show()
```

Now the chart shows the differences more accurately.

## Example 3: Using the Wrong Chart Type

Suppose we want to compare the number of students in different departments.

Using a line plot is not ideal because the categories are not continuous.

```python
departments = ["Physics", "Chemistry", "Biology", "Math"]
students = [80, 65, 120, 90]

plt.plot(departments, students)

plt.title("Students per Department")
plt.show()
```

### Why is this not ideal?

Line charts suggest **continuous trends over time**, but departments are **separate categories**.

A **bar chart** is better for comparing categories.

## Using a Better Chart Type

```python
plt.bar(departments, students)

plt.title("Number of Students by Department")
plt.xlabel("Department")
plt.ylabel("Number of Students")

plt.show()
```

Now the comparison between departments is clearer.

## Example 4: Too Many Colors

Using too many colors can make a chart hard to read.

```python
values = [15, 20, 35, 30]
labels = ["A", "B", "C", "D"]

plt.bar(labels, values,
        color=["red", "blue", "green", "purple"])

plt.title("Example Data")
plt.show()
```

### Why is this not ideal?

Multiple colors can be distracting when they are not needed.

If the bars represent the same type of data, using **one color** is usually better.

```python
plt.bar(labels, values, color="steelblue")

plt.title("Category Values")
plt.xlabel("Category")
plt.ylabel("Value")

plt.show()
```

Now the chart is simpler and easier to read.

## Example 5: Hard-to-Read Plot

Sometimes plots become difficult to read because text overlaps or the figure is too small.

```python
cities = ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"]
population = [8.5, 4.0, 2.7, 2.3, 1.6]

plt.bar(cities, population)
plt.title("Population of Cities")
plt.show()
```

### What is the problem?

City names overlap and are difficult to read.

We can fix this by:

- Rotating labels
- Increasing figure size

## Improved Version

```python
plt.figure(figsize=(8,5))

plt.bar(cities, population)

plt.title("Population of Major U.S. Cities")
plt.xlabel("City")
plt.ylabel("Population (Millions)")

plt.xticks(rotation=45)

plt.show()
```

## Key Lessons

Good visualizations help people understand data quickly and accurately.

When creating plots, remember these principles:

### Always Include Context
- Clear title
- Axis labels
- Units if necessary

### Use the Right Chart Type
- Line charts for trends over time
- Bar charts for category comparisons
- Scatter plots for relationships

### Avoid Misleading Scales
- Start bar charts at zero when possible
- Use appropriate axis ranges

### Keep Visuals Simple
- Avoid unnecessary colors
- Avoid clutter

### Improve Readability
- Rotate labels if needed
- Increase figure size
- Use clear fonts and spacing
